```python
# Import all necessary Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import warnings
warnings.filterwarnings('ignore')
```

```python
train_data = pd.read_html("https://docs.google.com/spreadsheets/d/e/2PACX-1vRTK2NvcndgPX41Czu6Ft2Ho_nE-z50BgTqdzwFW0rsJ2nvyNLe2DoIg1COzUbgw80oaRBjfy5-WtFk/pubhtml")
set1 = train_data[0]
set1
```

```python
set1.to_csv("/content/drive/MyDrive/Colab_Notebook/Internship_Name/CodesOnBytes/File_Name/Training_dataset.csv")
train_data = pd.read_csv("/content/drive/MyDrive/Colab_Notebook/Internship_Name/CodesOnBytes/File_Name/Training_dataset.csv")
train_data
```

```python
trains = train_data.drop("Unnamed: 0", axis = 1)
trains.columns = trains.iloc[0]
train_data = trains.drop(0)
train_data
```

```python
train_dataset = train_data.dropna(subset = ['y'])
train_dataset
```

```python
x_train = train_dataset[["x"]].astype(float)
y_train = train_dataset[["y"]].astype(float)
```

```python
plt.scatter(x_train, y_train, s = 5)
plt.grid(True, color = "grey", linewidth = "0.4")
plt.xlabel('X', fontname = 'serif', fontsize = 14)
plt.xticks(fontname = 'serif')
plt.ylabel('Y', fontname = 'serif', fontsize = 14)
plt.yticks(fontname = 'serif')
plt.show()
```

```python
chart = sns.regplot(x = x_train, y = y_train, scatter_kws = {'s':30, 'facecolor':'white'}, line_kws = {'color':'red'})
plt.grid(True, color = "grey", linewidth = "0.4")
plt.xlabel('x', fontname = 'serif', fontsize = 14)
plt.xticks(fontname = 'serif')
plt.ylabel('y', fontname = 'serif', fontsize = 14)
plt.yticks(fontname = 'serif')
plt.title("The regression line of y on x", fontname = 'serif', fontsize = 16)
plt.show()
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
m = model.intercept_
c = model.coef_
m, c
```

```python
yhat = model.predict(x_train)
x_train
```

```python
x_train.shape
```

```python
x_train = x_train.squeeze()
yhat = yhat.squeeze()
```

```python
index = [i for i in range(1, 700)]
plt.plot(index, y_train , color = "blue" , linewidth = 2, linestyle = '-')
plt.plot(index, yhat , color = "yellow", linewidth = 1, linestyle = '-')
plt.grid(True, color = "grey", linewidth = "0.4")
plt.xlabel('Index', fontname = 'serif', fontsize = 14)
plt.xticks(np.arange(0, 330, step = 30), fontname = 'serif', rotation = 90)
plt.ylabel('y-values', fontname = 'serif', fontsize = 14)
plt.yticks(np.arange(0, 120, step = 20), fontname = 'serif')
plt.title("Actual vs. Predicted", fontname = 'serif', fontsize = 16)
plt.show()
```

```python
from sklearn import metrics
```

```python
print("MSE = ", metrics.mean_squared_error(yhat, y_train))
print("R^2 = ", metrics.r2_score(yhat, y_train))
```

```python
plt.scatter(y_train, yhat, s = 7, color = 'red')
plt.grid(True, color = "grey", linewidth = "0.4")
plt.xlabel('Y-Train', fontname = 'serif', fontsize = 14)
plt.xticks(fontname = 'serif')
plt.ylabel('Predicted Y', fontname = 'serif', fontsize = 14)
plt.yticks(fontname = 'serif')
plt.show()
```
