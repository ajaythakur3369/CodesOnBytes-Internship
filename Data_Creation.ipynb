# **Problem Statement -**

Create a CSV dataset using Python, Pandas, and any public API.

**Dataset -** https://www.binance.com/api/v3/ticker/24hr

# **Let's Begin !**

```python
# Import all necessary Libraries
import pandas as pd
import requests
import warnings
warnings.filterwarnings('ignore')
```

```python
# Call an API and retrieve data through the API
api_url = "https://www.binance.com/api/v3/ticker/24hr"
```

```python
response = requests.get(api_url)
```

```python
# To access the file, access the Drive from Colab
from google.colab import drive
drive.mount('/content/drive/')
```

```python
# Put an if-else statement on the called API
if response.status_code == 200:
    data = response.json()
    df = pd.DataFrame(data)

    # This will create a CSV file named "Binance_dataset.csv" in the current directory
    df.to_csv('/content/drive/MyDrive/Colab_Notebook/Internship_Name/CodesOnBytes/File_Name/Binance_dataset.csv', index = False)
    print("Your CSV dataset has been created successfully.")
else:
  print("Unable to create the CSV file. Please check your URL")
```
