```python
sns.countplot(x = 'type', data = df, width = 0.3)
plt.title('Count vs Type of shows')
plt.grid(True, color = "grey", linewidth = "0.4")
plt.xlabel('type', fontname = 'serif', fontsize = 14)
plt.xticks(fontname = 'serif')
plt.ylabel('count', fontname = 'serif', fontsize = 14)
plt.yticks(fontname = 'serif')
plt.title("Count vs. Type of shows", fontname = 'serif', fontsize = 16)
plt.show()
```

This graph is plotted on the given dataset, which is not clean (missing values and outliers have not been removed)

```python
plt.figure(figsize = (15, 6))
sns.countplot(x = new_data_2['country'], order = new_data_2['country'].value_counts().index[0:10], width = 0.5)
plt.grid(True, color = "grey", linewidth = "0.4")
plt.xlabel('country', fontname = 'serif', fontsize = 14)
plt.xticks(fontname = 'serif')
plt.ylabel('count', fontname = 'serif', fontsize = 14)
plt.yticks(fontname = 'serif')
plt.title("Country-wise content on Netflix", fontname = 'serif', fontsize = 16)
plt.show()
```

```python
movie_countries = new_data_2[new_data_2['type'] == 'Movie']
tv_show_countries = new_data_2[new_data_2['type'] == 'TV Show']
```

```python
tv_show_countries
```

```python
plt.figure(figsize = (15, 6))
sns.countplot(x = 'country', order = new_data_2['country'].value_counts().index[0:10], data = movie_countries, width = 0.5)
plt.grid(True, color = "grey", linewidth = "0.4")
plt.xlabel('country', fontname = 'serif', fontsize = 14)
plt.xticks(fontname = 'serif')
plt.ylabel('count', fontname = 'serif', fontsize = 14)
plt.yticks(fontname = 'serif')
plt.title("Top 10 countries producing movies on Netflix", fontname = 'serif', fontsize = 16)
plt.show()
```

```python
plt.figure(figsize = (15, 6))
sns.countplot(x = 'country', order = new_data_2['country'].value_counts().index[0:10], data = tv_show_countries, width = 0.5)
plt.grid(True, color = "grey", linewidth = "0.4")
plt.xlabel('country', fontname = 'serif', fontsize = 14)
plt.xticks(fontname = 'serif')
plt.ylabel('count', fontname = 'serif', fontsize = 14)
plt.yticks(fontname = 'serif')
plt.title("Top 10 countries producing TV shows on Netflix", fontname = 'serif', fontsize = 16)
plt.show()
```

```python
ds = new_data_2.copy()
```

```python
plt.figure(figsize = (15, 6))
sns.countplot(x = 'director', order = ds['director'].value_counts().index[0:20], data = ds, width = 0.6)
plt.grid(True, color = "grey", linewidth = "0.4")
plt.xlabel('director', fontname = 'serif', fontsize = 14)
plt.xticks(rotation = 90, fontname = 'serif')
plt.ylabel('count', fontname = 'serif', fontsize = 14)
plt.yticks(fontname = 'serif')
plt.title("Top 20 Director's Shows On Netflix", fontname = 'serif', fontsize = 16)
plt.show()
```

```python
plt.figure(figsize = (15, 6))
sns.countplot(x = 'rating',order = ds['rating'].value_counts().index[0:10], data = ds, width = 0.5)
plt.grid(True, color = "grey", linewidth = "0.4")
plt.xlabel('rating', fontname = 'serif', fontsize = 14)
plt.xticks(fontname = 'serif')
plt.ylabel('count', fontname = 'serif', fontsize = 14)
plt.yticks(fontname = 'serif')
plt.title("Ratings of Shows on Netflix vs. Count", fontname = 'serif', fontsize = 16)
plt.show()
```

```python
plt.figure(figsize = (15, 6))
sns.countplot(x = 'listed_in', order = ds['listed_in'].value_counts().index[0:20], data = ds, width = 0.6)
plt.grid(True, color = "grey", linewidth = "0.4")
plt.xlabel('listed_in', fontname = 'serif', fontsize = 14)
plt.xticks(rotation = 90, fontname = 'serif')
plt.ylabel('count', fontname = 'serif', fontsize = 14)
plt.yticks(fontname = 'serif')
plt.title("Top 20 Genres on Netflix", fontname = 'serif', fontsize = 16)
plt.show()
```
